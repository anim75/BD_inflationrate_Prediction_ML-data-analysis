```python
#for stationary, the adf test statistic must be less than critical values, hence not stationary
```

```python
#making data sationary, three ways : TIME SHIFT, LOG, SQUARE/CUBE ROOT
```

```python
def test_stationarity(dataFrame, var):
    dataFrame['rollMean']  = dataFrame[var].rolling(window=12).mean()
    dataFrame['rollStd']  = dataFrame[var].rolling(window=12).std()
    
    from statsmodels.tsa.stattools import adfuller
    adfTest = adfuller(dataFrame[var],autolag='AIC')
    stats = pd.Series(adfTest[0:4],index=['Test Statistic','p-value','#lags used','number of observations used'])
    print(stats)
    
    for key, values in adfTest[4].items():
        print('criticality',key,":",values)
        
    sns.lineplot(data=dataFrame,x=dataFrame.index,y=var)
    sns.lineplot(data=dataFrame,x=dataFrame.index,y='rollMean')
    sns.lineplot(data=dataFrame,x=dataFrame.index,y='rollStd')
```

```python
air_df=df[['InflationRateYOY']]
air_df.head()
```

```python
#time shift

air_df['shift']=air_df.InflationRateYOY.shift()
air_df['shiftDiff']=air_df.InflationRateYOY-air_df['shift']
air_df.head()
```

```python
test_stationarity(air_df.dropna(),'shiftDiff')
```

```python
#log
log_df=df[['InflationRateYOY']]
log_df['log']=np.log(log_df['InflationRateYOY'])
log_df.head()
```

```python
test_stationarity(log_df.dropna(),'log')
```

```python
#sqrt
sqrt_df=df[['InflationRateYOY']]
sqrt_df['sqrt']=np.sqrt(df['InflationRateYOY'])
sqrt_df.head()


```

```python
test_stationarity(sqrt_df.dropna(),'sqrt')
```

```python
#cbrt
cbrt_df=df[['InflationRateYOY']]
cbrt_df['cbrt']=np.cbrt(df['InflationRateYOY'])
cbrt_df.head()
```

```python
test_stationarity(cbrt_df.dropna(),'cbrt')
```

```python
#log&sqrt
log_df2=log_df[['InflationRateYOY','log']]
log_df2['log_sqrt']=np.sqrt(log_df['log'])
log_df2.head()
```

```python
airP=df[['InflationRateYOY']].copy(deep=True)
airP['firstDiff']=airP['InflationRateYOY'].diff()
airP['Diff12']=airP['InflationRateYOY'].diff(12)
```

```python
airP.head()
```

```python
#ARIMA
#ar=p (pacf)
#i=d (shift)
#ma=q (acf)

from statsmodels.tsa.arima.model import ARIMA
from statsmodels.graphics.tsaplots import plot_acf,plot_pacf
```

```python
plot_pacf(airP['firstDiff'].dropna(),lags=20)
```

```python
plot_acf(airP['firstDiff'].dropna(),lags=20)
```

```python
#p=1(PACF) | d=1(SHIFT BY 1) | q=1(ACF)
```

```python
train=airP[:round(len(airP)*90/100)]
test=airP[round(len(airP)*90/100):]
test.tail()
```

```python
model=ARIMA(train['InflationRateYOY'],order=(1,1,1))
model_fit=model.fit()
prediction=model_fit.predict(start=test.index[0],end=test.index[-1])
airP['arimaPred']=prediction
airP.tail()
```

```python

airP.dropna()
sns.lineplot(data=airP,x=airP.index,y='InflationRateYOY')
sns.lineplot(data=airP,x=airP.index,y='arimaPred')
```

```python
from sklearn.metrics import mean_squared_error
```

```python
#finding error

np.sqrt(mean_squared_error(test['InflationRateYOY'],prediction))
```

```python
from statsmodels.tsa.statespace.sarimax import SARIMAX
```

```python
plot_pacf(airP['firstDiff'].dropna(),lags=20);
plot_acf(airP['firstDiff'].dropna(),lags=20);
```

```python
model = SARIMAX(train['InflationRateYOY'],order=(1,1,1),seasonal_order=(1,1,1,9))
model_fit = model.fit()
prediction = model_fit.predict(start=test.index[0],end=test.index[-1])
airP['sarimaxPred'] = prediction
```

```python
airP.dropna()
sns.lineplot(data=airP,x=airP.index,y='InflationRateYOY')
sns.lineplot(data=airP,x=airP.index,y='sarimaxPred')
sns.lineplot(data=airP,x=airP.index,y='arimaPred')
```
